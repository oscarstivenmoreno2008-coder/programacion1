```python
lista = ["colombia", "venezuela", "ecuador", "argentina", "chile"]
print(lista)
tupla = tuple(lista)
print(tupla)
conjunto = set(lista)
print(conjunto)
diccionario = dict(enumerate(lista))
print(diccionario)
```
